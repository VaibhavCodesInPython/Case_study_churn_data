```python
from google.colab import files
import pandas as pd
import numpy as np
import tensorflow as tf
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
df = pd.read_csv('Artificial_Neural_Network_Case_Study_data.csv')
```

```python
df.head()
```

```python
x = df.iloc[:,3:-1].values
y = df.iloc[:,-1].values
```

```python
le = LabelEncoder()
x[:, 2] = le.fit_transform(x[:, 2])
```

```python
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [1])], remainder='passthrough')
x = np.array(ct.fit_transform(x))
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
ann = tf.keras.models.Sequential()
```

```python
ann.add(tf.keras.layers.Dense(units=128, activation='relu'))
```

```python
ann.add(tf.keras.layers.Dense(units=64, activation='relu'))
```

```python
ann.add(tf.keras.layers.Dense(units=1, activation='sigmoid'))
```

```python
ann.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
ann.fit(x_train, y_train, batch_size=32, epochs=50)
```

```python
y_pred_prob = ann.predict(x_test)
y_pred = (y_pred_prob > 0.5)

print(f"Model Accuracy: {accuracy_score(y_test, y_pred):.2%}")
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred))
```

```python
test_results = pd.DataFrame({'Actual': y_test, 'Churn_Probability': y_pred_prob.flatten()})
high_risk_customers = test_results.sort_values(by='Churn_Probability', ascending=False).head(10)

print("Top 10 retention candidates: - ")
print(high_risk_customers)
```
